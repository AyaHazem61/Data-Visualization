```python
tips = sns.load_dataset('tips')
tips.sample(20)
```

#### sns.countplot(x = 'col_name1' | y = 'col_name2' , data = pandasDataFrameName)

```python
sns.countplot(x = 'size', data = tips)
plt.show()
```

```python
sns.countplot(y = 'day', data = tips)
plt.show()
```

```python
sns.countplot(x = 'size' , hue = 'time', data = tips)
plt.show()
```

#### sns.scatterplot( x = 'col_name1', y = 'col_name2', data = dataframename', hue = 'col_name')

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', size = 'size')
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', hue_order = ['Yes', 'No'])
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', hue_order = ['No', 'Yes'])
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', hue_order = ['No', 'Yes'], palette = {'Yes' : 'black' , 'No' : 'red'})
plt.show()
```

```python
sns.countplot(x = 'smoker', hue = 'sex' , data = tips)
plt.show()
```

```python
sns.countplot(x = 'smoker', hue = 'sex' , data = tips , hue_order = ['Female', 'Male'], palette = {'Female' : 'pink', 'Male' : 'blue'})
plt.show()
```

## Ch2 

#### relational plots ( line , scatter) are between two quantetive values such as time, stock

```python
sns.relplot(kind = 'scatter', x = 'total_bill', data = tips, y = 'tip', col = 'smoker')
plt.show() 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, row = 'smoker', kind = 'scatter')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, col = 'smoker', row = 'time', kind = 'scatter')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter', col = 'day')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter', col = 'day', col_wrap = 2, col_order = ['Thur', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter', hue = 'size', hue_order = [0,1,2,3,4,5,6], size = 'size')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', hue_order = ['Yes', 'No'], palette = {'Yes' : 'black','No' : 'red'}, style = 'smoker')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips , alpha = 0.4)
plt.show()
```
